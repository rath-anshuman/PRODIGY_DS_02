```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('titanic_cleaned.csv')

# Display basic statistics of numerical columns
print("Basic statistics of numerical columns:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of passengers by class (Pclass)
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df, palette='viridis')
plt.title('Distribution of Passengers by Class (Pclass)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the distribution of passengers by gender (Sex)
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df, palette='viridis')
plt.title('Distribution of Passengers by Gender (Sex)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the survival rate based on class and gender
plt.figure(figsize=(10, 6))
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=df, kind='count', palette='viridis')
plt.suptitle('Survival Count Based on Class and Gender')
plt.show()
```

```python
# Explore the distribution of fares
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], bins=30, kde=True, color='coral')
plt.title('Distribution of Passenger Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python

# Exclude non-numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['number'])

# Explore the correlation between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
